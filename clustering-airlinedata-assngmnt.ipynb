```python
from sklearn.preprocessing import StandardScaler
scaler_data = StandardScaler()
scaler_data=scaler_data.fit_transform(data)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,30):
  kmean=KMeans(n_clusters=i,init="k-means++",random_state=421)
  kmean.fit(scaler_data)
  wcss.append(kmean.inertia_)
  
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,30),wcss)
plt.title('elbow graph')
plt.xlabel("number of cluster")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters=15,random_state=231)
kmeans.fit_predict(scaler_data)
kmeans.labels_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaler_data,kmeans.labels_)
```

```python
plt.scatter(data['Balance'],data['Bonus_miles'],c=kmean.labels_)

plt.xlabel("balance")
plt.ylabel("bonus_miles")
plt.show()
```

```python
data['cluster_new']=kmean.labels_
data['cluster_new'].value_counts()
```

```python
import scipy.cluster.hierarchy as shc
dendro=shc.dendrogram(shc.linkage(scaler_data,method="ward"))
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
model.fit(scaler_data)
model.labels_
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(scaler_data,model.labels_)
```

```python
plt.scatter(data['Balance'],data['Bonus_miles'],c=model.labels_)

plt.xlabel("balance")
plt.ylabel("bonus_miles")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
model1=DBSCAN(eps=3,min_samples=5)
model1.fit(scaler_data)
model1.labels_
```

```python
plt.scatter(data['Balance'],data['Bonus_miles'],c=model1.labels_)

plt.xlabel("balance")
plt.ylabel("bonus_miles")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(scaler_data,model1.labels_)
```
